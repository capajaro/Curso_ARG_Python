```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
red = pd.read_csv('https://raw.githubusercontent.com/capajaro/Curso_ARG_Python/main/winequality-red.csv', delimiter=';')
white = pd.read_csv('https://raw.githubusercontent.com/capajaro/Curso_ARG_Python/main/winequality-white.csv', delimiter=';')
```

```python
print('----'*50)
print('Column red wine dataset: ' + red.columns)
print('----'*50)
print('Column white wine dataset: ' + white.columns)
```

```python
mean_red = red.mean()
mean_white = white.mean()
mean_comp = mean_white-mean_red
```

```python
print('----'*50)
Val_Lis = mean_red.values
VaL_Tag = ['%0.3f'%(el) for el in Val_Lis]
print('\u00b5 of '+ mean_red.index + ' en base de datos de vino rojo: '+ VaL_Tag)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(6,5))
plt.plot(mean_red, label = '$\mu$ red wine', color ='darkred')
plt.plot(mean_white, label = '$\mu$ white wine', color ='k')
plt.plot(mean_comp, label = 'comparison $\mu$ ', color ='blue')

plt.xticks(rotation = 90)

plt.title('Comparación Media de variables', fontdict={
  'fontsize': 16, 'fontname': 'serif', 'fontweight': 'bold'})
plt.legend()
ax =plt.gca()
ax.set_xticklabels(labels = mean_red.index , fontdict={
  'fontsize': 12, 'fontname': 'serif'})
ax.set_yticks(ticks = np.arange(0,160,20))
ax.set_yticklabels(labels = np.arange(0,160,20) , fontdict={
  'fontsize': 12, 'fontname': 'serif'})
plt.grid(b=True, which='both', axis='both')
plt.legend(prop = {'family':'serif', 'size':10})
plt.tight_layout()
```
